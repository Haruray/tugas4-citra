```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from PIL import Image
import cv2

from sklearn.model_selection import GridSearchCV 
from sklearn.model_selection import train_test_split 
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report

seed = 1
random.seed(seed)
np.random.seed(seed)
```

```python
IMG_SIZE = 220
categories=['Bus','Car', 'Truck', 'motorcycle'] 
training_data=[]
train_dir='/kaggle/input/vehicle-type-recognition/Dataset/' 
#path which contains all the categories of images 
for category in categories: 
    print(f'loading... category : {category}') 
    # Load train data
    path=os.path.join(train_dir,category) 
    class_num=categories.index(category)
    for img in os.listdir(path):
        img_array=cv2.imread(os.path.join(path,img)) 
        img_resized=cv2.resize(img_array,(IMG_SIZE, IMG_SIZE))
#         img_resized = cv2.cvtColor(img_resized, cv2.COLOR_BGR2GRAY)
        training_data.append([img_resized,class_num])
    print(f'loaded category for train:{category} successfully') 
```

```python
train_data_len = len(training_data)
```

```python
X_train=[]
y_train=[]

for categories, label in training_data:
    X_train.append(categories)
    y_train.append(label)
X_train = np.array(X_train).reshape(train_data_len,-1)
y_train = np.array(y_train)
```

```python
X_train = X_train/255.0
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_train,y_train, test_size=0.1)
```

```python
from sklearn.svm import SVC
svc = SVC(kernel='linear',gamma='auto',probability=True)
# model=GridSearchCV(svc,param_grid)
# model.fit(X_train,y_train)
svc.fit(X_train, y_train)
```

```python
y2 = svc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy on unknown data is",accuracy_score(y_test,y2))
```
